#### 2. Задание (на листочке)
Почему прямые не кажутся перпендикулярными? (см.ролик)

```python
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2 = (-1/3) * x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
```

В аналитическом выражении прямые, заданные линейными функциями y = k*x + b и y = a*x + m будут перпендикулярны, если выполнено следующее условие на их угловые коэффициенты: a*k = -1

```python
print(3 * (-1/3))
```

Условие на угловые коэффициенты выполнено, следовательно заданные в условии прямые перпендикулярны.

Прямые не кажутся перпендикулярными, так как на графике у осей x и y разные пределы.
Установим на обоих осях одинаковые пределы значений (-15, 15) и посмотрим на результат.

```python
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2 = (-1/3) * x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlim([-15, 15])
plt.ylim([-15, 15])
plt.xlabel('x')
plt.ylabel('y')
```

После установления на осях одинаковых пределов x и y, прямые на графике выглядят перепендикулярно друг к дургу.

#### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python
def circle(A=[0, 0], R=5):
    """
    A - center
    R - radius
    """
    a, b = A
    x = np.linspace((a - R), (a + R), 121)
    y1 = np.sqrt(np.power(R, 2) - np.power((x - a), 2)) + b
    y2 = -1 * np.sqrt(np.power(R, 2) - np.power((x - a), 2)) + b
    plt.plot(x, y1, color='blue')
    plt.plot(x, y2, color='blue')
    plt.xlim([(a - R - 1), (a + R + 1)])
    plt.ylim([(b - R - 1), (b + R + 1)])
    plt.show()
```

```python
circle()
```

```python
def ellipse(A=[0, 0], a=10, b=5):
    """
    A - center
    a, b - major and minor axis
    """
    x0, y0 = A
    x = np.linspace((x0 - a), (x0 + a), 121)
    y1 = np.sqrt(np.power(b, 2) - np.power(b, 2) *  np.power((x - x0), 2) / np.power(a, 2)) + y0
    y2 = -1 * np.sqrt(np.power(b, 2) - np.power(b, 2) *  np.power((x - x0), 2) / np.power(a, 2)) + y0
    plt.plot(x, y1, color='blue')
    plt.plot(x, y2, color='blue')
    plt.xlim([(x0 - a - 1), (x0 + a + 1)])
    plt.ylim([(y0 - b - 1), (y0 + b + 1)])
    plt.show()
```

```python
ellipse()
```

```python
def hyperbola(x=np.linspace(-5, 5, 121), a=1, b=1):
    y1 = np.sqrt((x * b / a - b) * ((x * b / a + b)))
    y2 = -1 * np.sqrt((x * b / a - b) * ((x * b / a + b)))
    plt.plot(x, y1, color='blue')
    plt.plot(x, y2, color='blue')
    plt.show()
```

```python
hyperbola()
```
