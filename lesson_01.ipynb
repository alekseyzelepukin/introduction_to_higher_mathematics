#### 4. Задание (в программе):
Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-5, 5, 121)
```

```python
def y(k, x):
    return np.cos(k*x)
```

```python
plt.plot(x, y(1, x), label='k=1', color='blue')
plt.plot(x, y(4, x), label='k=4', color='red')
plt.xlabel('x')
plt.ylabel('y(k,x)=cos(k*x)')
plt.legend()
```
