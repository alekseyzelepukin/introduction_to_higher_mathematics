#### 1. Задание (в программе)
Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-np.pi, np.pi, 201)
```

```python
def y(x, k, a, b):
    return k*np.cos(x-a)+b
```

```python
k, a, b = 1, 2, 3
plt.plot(x, y(x, k, a, b), label=f'k={k}, a={a}, b={b}', color='blue')
k, a, b = 10, 5, 0
plt.plot(x, y(x, k, a, b), label=f'k={k}, a={a}, b={b}', color='red')
k, a, b = 4, 12, 8
plt.plot(x, y(x, k, a, b), label=f'k={k}, a={a}, b={b}', color='green')
plt.xlabel('x')
plt.ylabel('y(x)=k*cos(x-a)+b')
plt.grid(True)
plt.legend()
```

#### 2. Задание 
Докажите, что при ортогональном преобразовании сохраняется расстояние между точками.
Нарисуйте график окружности в полярных координатах.

```python
x, y = 1, 3
a = np.pi / 4 
sin_a = np.sin(a)
cos_a = np.cos(a)
x1, y1 = x*sin_a + y*cos_a, x*-sin_a + y*cos_a
plt.plot(x, y, color='blue', marker='o')
plt.plot(x1, y1, color='red', marker='o')
```

```python
norm = np.round(np.linalg.norm([x, y]), 4)
norm1 = np.round(np.linalg.norm([x1, y1]), 4)
print(norm, norm1)
```

```python
print(norm == norm1)
```

Как мы видим, при линейном преобразовании расстояние между точками сохранилось.
Докажем, что преобразование является ортогональным.

```python
# условия ортогональности линейного преобразования:
# a11^2 + a12^2 = 1
# a21^2 + a22^2 = 1
# a11*a12 + a21*a22 = 0
# a11*a22 - a12*a21 != 0

# проверка:
a11, a12, a21, a22 = sin_a, cos_a, -sin_a, cos_a

print(f'a11^2 + a12^2 = 1: {np.round(a11**2 + a12**2, 1) == 1.0}')
print(f'a21^2 + a22^2 = 1: {np.round(a21**2 + a22**2, 1) == 1.0}')
print(f'a11*a12 + a21*a22 = 0: {np.round(a11*a12 + a21*a22, 1) == 0.0}')
print(f'a11*a12 + a21*a22 = 0: {np.round(a11*a22 - a12*a21, 1) != 0.0}')
```

Из вычислений выше, мы видим, что наше линейное преобразование точек является ортогональным. Так как расстояние между точкамисохранились, то мы можем сделать вывод, что при ортогональным преобразование расстояние между точками сохраняется. 

```python
R = 1
phi = np.linspace(0, 2 * np.pi, 201)
r = np.linspace(R, R, 201)
plt.polar(phi, r, color='green')
```

#### 3. Задание (в программе)
Напишите код, который будет переводить полярные координаты в декартовы.
Напишите код, который будет рисовать график окружности в полярных координатах.
Напишите код, который будет рисовать график прямой линии в полярных координатах.

```python
def polar_to_cartesian(R, a):
    """
    R - radius
    a - angle in degrees
    """
    x = R * np.cos(np.radians(a))
    y = R * np.sin(np.radians(a))
    return x, y
```

```python
print(polar_to_cartesian(10, 60))
```

```python
def polar_circle(R=1):
    phi = np.linspace(0, 2 * np.pi, 201)
    r = np.linspace(R, R, 201)
    plt.polar(phi, r, color='blue')
    plt.show()
```

```python
polar_circle()
```

```python
def polar_line(R=1, a=45):
    phi = np.array([np.radians(a), np.radians(a)])
    r = np.array([0, R])
    plt.polar(phi, r)
    plt.show()
```

```python
polar_line(R=5, a=30)
```

#### 4. Задание (на листочке)
Решите систему уравнений:  
y = x2 – 1  
exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve
```

```python
def equitations(p):
    x, y = p
    return y - x**2 + 1, np.exp(x) + x*(1-y) - 1

x1, y1 = fsolve(equitations, (0, 0))

print(x1, y1)
```

Решите систему уравнений и неравенств:  
y = x2 – 1  
exp(x) + x∙(1 – y) > 1

```python
# подберем начальную точку для решения уравнения, которая удовлетворяет неравенству
x = 0
while True:
    y = x**2 - 1
    if np.exp(x) + x*(1-y) > 1:
        print(f'x: {x}, y: {y}')
        break
    else:
        x += 1
```

```python
# проверка:
print(np.exp(x) + x*(1-0) - 1 > 0)
```

```python
# подставим найденную точку для поиска корней уравнения
x2, y2 = fsolve(equitations, (1, 0))

print(x2, y2)
```
